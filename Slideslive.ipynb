```python
# modules
import re
import requests

from bs4 import BeautifulSoup
# from selenium import webdriver
```

```python
# url = 'https://slideslive.com/38972123/mix3d-outofcontext-data-augmentation-for-3d-scenes'# yoda
url = "https://slideslive.com/38917998/assisting-vulnerable-communities-through-ai-and-or-from-data-to-deployed-decisions?ref=account-32515-presentations"
html_text = requests.get(url).text
soup = BeautifulSoup(html_text, 'html.parser')
print(soup.prettify())
```

```python
html_text = requests.get(url).text
soup = BeautifulSoup(html_text, 'html.parser')
video_id = url.split("https://slideslive.com/")[-1].split("/")[0]
vid_slide = soup.find("div", {"data-mode": "video_slideshow"})
if vid_slide:
    token = soup.find("div", {"data-mode": "video_slideshow"})["data-player-token"]
    json_url = f"https://ben.slideslive.com/player/{video_id}?player_token={token}"
    print(json_url)
else:# ignore if no video-slide architecture like pannel discussions
    print(f"{url}: video-silde architecture not detected")
```

```python
html_text = requests.get(json_url).text
soup = BeautifulSoup(html_text, 'html.parser')
print(soup.prettify())
```

```python
service_name = html_text.split("#EXT-SL-VOD-VIDEO-SERVICE-NAME:")[1].split("\n")[0]
if service_name=="youtube":
    video_id = html_text.split("#EXT-SL-VOD-VIDEO-ID:")[1].split("\n")[0]
else:
    video_id = None
    print(f"Service name is {service_name}")
```

```python
if video_id:
    download_url = f"https://www.youtube.com/watch?v={video_id}"
else:
    print("Not sure about how to download")
```
